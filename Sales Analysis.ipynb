```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df[""].value_counts()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Discounted Sales'] = df['Total Sales'] * 0.95
```

```python
df
```

```python
def categorize_sales(sales):
    if sales < 500:
        return "Low"
    elif 500 <= sales < 1500:
        return "Medium"
    else:
        return "High"
df['Sales Category'] = df['Total Sales'].apply(categorize_sales)
```

```python
df
```

```python
df['Day of Week'] = df['Date'].dt.day_name()
```

```python
df
```

```python
df['Week of Year'] = df['Date'].dt.isocalendar().week
```

```python
df
```

```python
df['Quarter'] = df['Date'].dt.quarter
```

```python
df
```

```python
Q1 = df['Total Sales'].quantile(0.25)
Q3 = df['Total Sales'].quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers = df[(df['Total Sales'] < (Q1 - 1.5 * IQR)) | (df['Total Sales'] > (Q3 + 1.5 * IQR))]
print(f"Number of Outliers in Total Sales: {len(outliers)}")
```

```python
print(df.describe())
```

```python
sales_by_product = df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
print(sales_by_product)
```

```python
sales_by_region = df.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
print(sales_by_region)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values, palette='viridis')
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Total Sales by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=sales_by_region.index, y=sales_by_region.values, palette='coolwarm')
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.title("Total Sales by Region")
plt.show()
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total Sales'].sum()
plt.figure(figsize=(12, 5))
monthly_sales.plot(marker='o', color='b')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.grid()
plt.show()
```
